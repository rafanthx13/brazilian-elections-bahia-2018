```python
def to_df_from_list_of_objs_sorted(list_of_obj):
    """
        Converte de:
    [{'id': 5, 'nome': 'rafael'},  {'id': 6, 'nome': 'marta'}]
        Para DataFrame
    OBS: Esse processo é mais complxo pois tem que ser feito
    """
    variables = list_of_obj[0].keys()

    def aux_obj_to_list_sorted(obj, features):
        alist = []
        for feat in features:
            alist.append(obj[feat])
        return alist

    return pd.DataFrame(
                list( map( lambda x: aux_obj_to_list_sorted(x, variables), list_of_obj)),
                columns = variables)
```

```python
def return_df_from_json_file(path_file, produto):
    start_path = './json-files/'
    full_path = start_path + path_file
    files = [full_path + f for f in listdir(full_path) if isfile(join(full_path, f))]
    great_df = []
    for file_get in files:
        json_data = json.load(open(file_get))
        estabelecimentos = [
            'RONDELLI COMERCIO E TRANSPORTE LTDA',
            'CAMBUI SUPERMERCADOS',
            'SACOLA CHEIA',
            'MINEIRAO ATACAREJO',
            'O ATACADAO',
            'SUPERMERCADO CARDOSO',
        ]
        date_query = json_data['dataConsulta']
        list_objs = []
        for el in json_data['resultado']:
            if(el['estabelecimento']['nomeEstabelecimento'] in estabelecimentos):
                adict = {}
                adict['produto'] = produto
                adict['produto_nome'] = el['produto']['descricao']
                adict['unidade'] = el['produto']['unidade']
                adict['date_query'] = date_query
                adict['estabelecimento'] = el['estabelecimento']['nomeEstabelecimento']
                adict['rua'] = el['estabelecimento']['endLogradouro']+ ' - '+el['estabelecimento']['bairro']
                adict['preco'] = el['produto']['precoUnitario']
                adict['bairro'] = el['estabelecimento']['bairro']
                list_objs.append(adict)
        # Formating
        temp_df =  to_df_from_list_of_objs_sorted(list_objs)
        temp_df['date_query'] = pd.to_datetime(temp_df["date_query"])
        temp_df['date_formated'] = temp_df["date_query"].dt.strftime('%d/%m/%Y %H:%M')
        temp_df['estabelecimento'] = temp_df['estabelecimento'].replace({
            'RONDELLI COMERCIO E TRANSPORTE LTDA': 'RONDELLI',
            'CAMBUI SUPERMERCADOS': 'CAMBUI',
        })
        # Format estabelcimento pondo bairro
        list_change_name = [
            'RONDELLI', 'SUPERMERCADO CARDOSO'
        ]
        for i, row in alfa.iterrows():
            if(row['estabelecimento'] in list_change_name):
                alfa.loc[i, 'estabelecimento'] = row['estabelecimento'] + ' - ' + row['bairro']
        
        great_df.append(temp_df)
    return pd.concat(great_df, ignore_index=True)
```
